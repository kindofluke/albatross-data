## Benchmark Function & Execution

### Subtask:
Create a function to execute queries on both DuckDB and data_kernel, measure execution time, and run it for all defined queries.

**Reasoning**:
The previous subtask of defining the queries is complete. Now I will create the benchmarking function and execute the queries.



```python
def run_benchmark(query_info):
    name = query_info['name']
    duckdb_sql = query_info['duckdb_sql']
    data_kernel_sql = query_info['data_kernel_sql']

    # DuckDB
    start_time = time.time()
    try:
        duckdb.sql(duckdb_sql).df()
        duckdb_time = time.time() - start_time
    except Exception as e:
        print(f"Error running DuckDB for {name}: {e}")
        duckdb_time = None
    print("Run the GPUS")
    # Data Kernel
    start_time = time.time()
    try:
        # Handling the specific data_kernel API
        data_kernel.execute(data_kernel_sql)['result'].to_list()
        dk_time = time.time() - start_time
    except Exception as e:
        print(f"Error running Data Kernel for {name}: {e}")
        dk_time = None

    return {
        'Query': name,
        'DuckDB (CPU)': duckdb_time,
        'data_kernel (GPU)': dk_time
    }

results = []
for q in queries:
    print(f"Running {q['name']}...")
    res = run_benchmark(q)
    results.append(res)

df_results = pd.DataFrame(results)
print("Benchmarking complete.\n")
print(df_results.to_markdown(index=False))
```

## Visualize Performance Results with Altair

### Subtask:
Visualize the performance results using Altair to compare DuckDB and data_kernel execution times.


**Reasoning**:
I will create a code block to melt the `df_results` DataFrame and generate a grouped bar chart using Altair to compare the execution times of DuckDB and data_kernel across all queries.



```python
import altair as alt

# Ensure df_results is defined by running the previous cell first.
# If it is defined, melt the dataframe for Altair
if 'df_results' in locals():
    df_melted = df_results.melt(
        id_vars=['Query'],
        value_vars=['DuckDB (CPU)', 'data_kernel (GPU)'],
        var_name='Engine',
        value_name='Execution Time (s)'
    )

    # Create the grouped bar chart using xOffset
    chart = alt.Chart(df_melted).mark_bar().encode(
        x=alt.X('Query:N', title='Query', axis=alt.Axis(labelAngle=-45)),
        y=alt.Y('Execution Time (s):Q', title='Execution Time (seconds)'),
        xOffset='Engine:N',
        color=alt.Color('Engine:N', title='Engine')
    ).properties(
        title='DuckDB (CPU) vs data_kernel (GPU) Performance',
        width=600,
        height=400
    )

    display(chart)
else:
    print("Please run the previous cell to generate 'df_results' before running this cell.")
```
